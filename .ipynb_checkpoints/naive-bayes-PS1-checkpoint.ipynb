```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

There are 10 types of faults, linked to each bearing defect:

- Ball_007_1: Ball defect (0.007 inch)
- Ball_014_1: Ball defect (0.014 inch)
- Ball_021_1: Ball defect (0.021 inch)
- IR_007_1: Inner race fault (0.007 inch)
- IR_014_1: Inner race fault (0.014 inch)
- IR_021_1: Inner race fault (0.021 inch)
- Normal_1: Normal
- OR_007_6_1: Outer race fault (0.007 inch, data collected from 6 O'clock position)
- OR_014_6_1: Outer race fault (0.014 inch, 6 O'clock)
- OR_021_6_1: Outer race fault (0.021 inch, 6 O'clock)

Reading the data into a Pandas dataframe,

```python
data = pd.read_csv("feature_time_48k_2048_load_1.csv")
data
```

split the datasets in a 3:1 ratio (train/test)


```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data, test_size = 575, stratify = data['fault'], random_state = 1234)
print( train_data['fault'].value_counts(), "\n")
print( test_data['fault'].value_counts(), "\n")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train_data_scaled = scaler.fit_transform(train_data.iloc[:,:-1]) # Skip last column 'fault'
test_data_scaled = (test_data.iloc[:,:-1].values - scaler.mean_)/np.sqrt(scaler.var_)

pd.DataFrame(test_data_scaled).describe()
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import mean_squared_error
nb_model = GaussianNB(var_smoothing = 1e-05)

# the model
nb_model.fit(train_data_scaled, train_data['fault'])
```

Actually making predictions,

```python
from sklearn.metrics import confusion_matrix,classification_report, accuracy_score
```

```python
train_predictions = nb_model.predict(train_data_scaled)
test_predictions = nb_model.predict(test_data_scaled)
```

Plot confusion matrixes:

```python
fault_type = data.fault.unique()

test_confu_matrix = confusion_matrix(test_data['fault'], test_predictions)
plt.figure(1,figsize=(18,8))
plt.subplot(122)
sns.heatmap(test_confu_matrix, annot = True,
xticklabels=fault_type, yticklabels=fault_type, cmap = "Blues", cbar = False)
plt.title('Test Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')

plt.show()
```

```python
print("Accuracy:", accuracy_score(test_data['fault'], test_predictions))
```

```python
class_report = classification_report(y_pred = test_predictions, y_true = test_data['fault'])
print(class_report)
```

# To compare with the MPFT dataset results

Since the CWRU dataset has 10 possible classes, that could be a reason to explain its lower accuracy score compared to the MPFT dataset's accuracy.

Now, we standardise the nine types of faults to two types (inner race faults, and outer race faults) to compare that specific data with the MPFT results

```python
data_tocmp = pd.read_csv("feature_time_48k_2048_load_1.csv")

data_tocmp['fault'] = data_tocmp['fault'].replace(['Ball_007_1','Ball_014_1','Ball_021_1'],'Ball fault')
data_tocmp['fault'] = data_tocmp['fault'].replace(['IR_007_1','IR_014_1','IR_021_1'],'Inner race fault')
data_tocmp['fault'] = data_tocmp['fault'].replace(['OR_007_6_1','OR_014_6_1','OR_021_6_1'],'Outer race fault')
data_tocmp['fault'] = data_tocmp['fault'].replace(['Normal_1'],'Normal')

data_tocmp = data_tocmp[data_tocmp['fault']!='Ball fault']



data_tocmp
```

```python
train_data, test_data = train_test_split(data_tocmp, test_size = 575, stratify = data_tocmp['fault'], random_state = 1234)
train_data_scaled = scaler.fit_transform(train_data.iloc[:,:-1])
test_data_scaled = (test_data.iloc[:,:-1].values - scaler.mean_)/np.sqrt(scaler.var_)

nb_model = GaussianNB(var_smoothing = 1e-05)
nb_model.fit(train_data_scaled, train_data['fault'])

train_predictions = nb_model.predict(train_data_scaled)
test_predictions = nb_model.predict(test_data_scaled)

fault_type = data_tocmp.fault.unique()

test_confu_matrix = confusion_matrix(test_data['fault'], test_predictions)
plt.figure(1,figsize=(18,8))
plt.subplot(122)
sns.heatmap(test_confu_matrix, annot = True,
xticklabels=fault_type, yticklabels=fault_type, cmap = "Blues", cbar = False)
plt.title('Test Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')

plt.show()


print("Accuracy:", accuracy_score(test_data['fault'], test_predictions))
class_report = classification_report(y_pred = test_predictions, y_true = test_data['fault'])
print(class_report)
```

```python
import random
list =[0]*575

print("Final results\n" )

values= test_data['fault'].values
for i in range (len(test_predictions)):
    list[i]=[test_predictions[i],values[i],str(test_predictions[i]==values[i])]

random.shuffle(list)
pd.DataFrame(list,columns=['Predicted Fault','Actual Fault','Correct?'])
```
